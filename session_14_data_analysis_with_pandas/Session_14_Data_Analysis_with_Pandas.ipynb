```python
# We can get the Latitude and Logintude from the Geolocation
n = nom.geocode("3995 23rd St, San Francisco, CA 94114")
# print(n)
print("Latitude: {} and Longitude: {}".format(n.latitude,n.longitude))
```

```python
import pandas
df9 = pandas.read_csv("supermarkets.csv")
df9
```

```python
# Changing the Address Columm with more values
df9["Address"] = df9["Address"] + ", " + df9["City"] + ", " + df9["State"] + ", " + df9["Country"]
df9
```

```python
#
df9["Coordinates"] = df9["Address"].apply(nom.geocode)
df9.Coordinates
```

```python
df9.Coordinates[5].latitude
```

```python
df9["Latitude"] = df9["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df9["Longitude"] = df9["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df9
```
